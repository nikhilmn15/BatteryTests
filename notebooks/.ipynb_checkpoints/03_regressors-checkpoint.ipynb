# SOH & RUL Regressors -- MIT Battery Dataset

The two regression components of the pipeline. Covers algorithm comparison, why Gradient Boosting won both, feature importances, and honest predicted-vs-true plots on genuinely held-out batteries.

Companion to `01_eda.ipynb` / `02_feature_engineering.ipynb`. Outputs pre-baked from a real run.

## 1. Setup

```python
import sys
sys.path.append('../pipeline_mit')
import numpy as np
import matplotlib.pyplot as plt
import models

df = models.load_clean_data()
data = models.get_model_data(df)
print(f"{len(data)} rows, {data.battery_id.nunique()} batteries")
print("Model features:", models.MODEL_FEATURES)
```

## 2. Algorithm comparison -- GroupKFold, never random split

Grouped by `battery_id` so no battery's cycles leak across train/test. This matters enormously here -- a random split would let the model "cheat" by seeing cycle 500 of a battery whose cycle 501 is in the test set (nearly identical), producing a misleadingly optimistic score.

```python
soh_results, soh_best = models.train_soh(df)
```

```python
rul_results, rul_best = models.train_rul(df)
```

**Why Gradient Boosting wins both, and why Linear Regression is so far behind (especially RUL):** the relationship between `chargetime_ratio`/`IR_ratio` and remaining life almost certainly isn't linear -- degradation accelerates non-linearly as a cell nears end-of-life (the same "knee point" phenomenon seen in the NASA dataset). Tree-based ensembles capture that curvature naturally; a linear model is structurally unable to, regardless of how much data it sees.

## 3. Feature importance -- which signals actually drive each prediction

```python
import joblib
soh_bundle = joblib.load('../models_mit/soh_model.pkl')
rul_bundle = joblib.load('../models_mit/rul_model.pkl')

print("SOH feature importances:")
for f, imp in sorted(zip(soh_bundle['features'], soh_bundle['model'].feature_importances_), key=lambda x: -x[1]):
    print(f"  {f:18s} {imp:.4f}")
```

```python
print("RUL feature importances:")
for f, imp in sorted(zip(rul_bundle['features'], rul_bundle['model'].feature_importances_), key=lambda x: -x[1]):
    print(f"  {f:18s} {imp:.4f}")
```

**`chargetime_ratio` dominates both tasks** (60% of SOH's decision-making, 49% of RUL's) -- consistent with the correlation analysis in the feature engineering notebook (chargetime_ratio: -0.697 with soh). `IR_ratio` is a strong second signal in both. Note RUL leans more on the protocol parameters (`Q1`, `C2`) than SOH does -- makes some sense, since which charging protocol a battery uses is itself informative about how much life it's likely to have left, independent of current degradation state.

## 4. Predicted vs. true -- on a genuinely held-out battery group

```python
from sklearn.model_selection import GroupKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor

X = data[models.MODEL_FEATURES].values
groups = data['battery_id'].values
gkf = GroupKFold(n_splits=5)
tr, te = next(gkf.split(X, data['soh'].values, groups))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11,5))

scaler = StandardScaler(); Xtr, Xte = scaler.fit_transform(X[tr]), scaler.transform(X[te])
y_soh = data['soh'].values
m = GradientBoostingRegressor(n_estimators=100, max_depth=3, random_state=42)
m.fit(Xtr, y_soh[tr]); pred_soh = m.predict(Xte)
ax1.scatter(y_soh[te], pred_soh, s=2, alpha=0.15, color='#2a78d6')
ax1.plot([0.7,1.02],[0.7,1.02], color='#898781', linestyle='--', linewidth=1)
ax1.set_xlabel('True SOH'); ax1.set_ylabel('Predicted SOH')
ax1.set_title('SOH: predicted vs true (held-out fold)')
ax1.grid(alpha=0.3)

y_rul = data['rul_cycles'].values
m2 = GradientBoostingRegressor(n_estimators=100, max_depth=3, random_state=42)
m2.fit(Xtr, y_rul[tr]); pred_rul = m2.predict(Xte)
ax2.scatter(y_rul[te], pred_rul, s=2, alpha=0.15, color='#1baf7a')
ax2.plot([0,2200],[0,2200], color='#898781', linestyle='--', linewidth=1)
ax2.set_xlabel('True RUL (cycles)'); ax2.set_ylabel('Predicted RUL (cycles)')
ax2.set_title('RUL: predicted vs true (held-out fold)')
ax2.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

**Reading these honestly, not just admiring the diagonal trend:**
- SOH's scatter hugs the diagonal reasonably tightly -- consistent with R2=0.89.
- RUL's scatter is visibly noisier, especially at low true-RUL values (bottom-left) -- exactly where accuracy matters most for a real replacement decision, and exactly where R2=0.62 (vs SOH's 0.89) should set your expectations. A predicted RUL of "200 cycles" here could plausibly mean anywhere from ~50 to ~350 in reality -- useful as a rough signal, not a number to plan a maintenance schedule around precisely.

## 5. Summary

- Gradient Boosting won both tasks, and by a wide margin over Linear Regression on RUL specifically -- non-linear "knee point" degradation is real in this dataset, not just the NASA one.
- `chargetime_ratio` is the single most important feature for both tasks -- validates the feature engineering decisions made in notebook 02.
- SOH (R2=0.89) is meaningfully more reliable than RUL (R2=0.62) -- worth remembering when the agent reports both with equal confidence in its phrasing.
